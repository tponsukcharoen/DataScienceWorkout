Create a master data frame.

```python
BalanceSheet = pd.DataFrame()
IncomeStatement = pd.DataFrame()
CashFlowStatement = pd.DataFrame()
```

```python
filename = "Financial_Report_2010.xlsx"
xl = pd.ExcelFile(DataDir+filename)
#xl.sheet_names
```

```python
d = xl.parse('CONSOLIDATED_BALANCE_SHEET')
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d = d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
BalanceSheet = d
d = xl.parse('CONSOLIDATED_STATEMENT_OF_INCO')
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d = d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
IncomeStatement = d
d = xl.parse('CONSOLIDATED_STATEMENT_OF_CASH')
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d= d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
CashFlowStatement = d
```

```python
d
```

```python
filename = "Financial_Report_2012.xlsx"
xl = pd.ExcelFile(DataDir+filename)
#xl.sheet_names
```

```python
d = xl.parse('Consolidated_Balance_Sheet')
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d = d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
BalanceSheet = pd.merge(BalanceSheet, d, how='outer', on=["metric"])

d = xl.parse('Consolidated_Statement_of_Inco') 
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d = d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
IncomeStatement = pd.merge(IncomeStatement, d, how='outer', on=["metric"])

d = xl.parse('Consolidated_Statement_of_Cash')
d = d.rename(index=str, columns={d.columns[0]: "metric"})
d = d.ix[:,[d.columns[0],d.columns[2],d.columns[1]]]
d = d.reset_index()
CashFlowStatement = pd.merge(CashFlowStatement, d, how='outer', on=["metric"])
```
